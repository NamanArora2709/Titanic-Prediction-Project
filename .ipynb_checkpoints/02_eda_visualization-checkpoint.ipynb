```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the same Titanic dataset (already cleaned if you're sharing notebook)
df = pd.read_csv('train.csv')  # OR skip if working in same notebook as Person 1

# (Optional) Clean again if working independently
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop(['Cabin', 'Ticket', 'PassengerId', 'Name'], axis=1)
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)

# Step 3: Plot survival count
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()

# Step 4: Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.xticks([0, 1], ['Male', 'Female'])
plt.show()

# Step 5: Survival by Passenger Class
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Class')
plt.show()

# Step 6: Survival by Age (Histogram)
sns.histplot(data=df, x='Age', bins=20, hue='Survived', kde=True)
plt.title('Survival by Age')
plt.show()

# Step 7: Survival by Age Group

# Create AgeGroup column
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 50, 80], 
                        labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])

# Plot Survival by Age Group
plt.figure(figsize=(8, 5))
sns.countplot(x='AgeGroup', hue='Survived', data=df, palette='Set2')
plt.title('Survival by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()

# Optional: Remove AgeGroup after plotting if not needed
df.drop('AgeGroup', axis=1, inplace=True)


```
